```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math
```

```python
N = 4
mat = Matrix.zeros(N, N)

for i in range(N):
    for j in range(N):
        mat[i,j] = symbols('a%d%d'%(i+1, j+1))

display(Math(r'A = %s'%latex(mat)))
```

```python
tr = []
A = Matrix.eye(N)

for i in range(N):
    for j in range(N):
        A = mat * A
        tr.append(Matrix.trace(A))

for i in range(N):
    tr[i] = tr[i].simplify()
    display(Math(r'trace(A^{%d}) = %s'%(i + 1, latex(tr[i]))))
```

```python
F = Matrix.zeros(N, 1)
R = Matrix.zeros(N + 1, 1)

R[N] = 1
F[0] = tr[0]
R[N - 1] = -tr[0]
FT = -1
for i in range(1, N):
    F[i] = 0
    P = 1
    for j in range(i):
        F[i] += P * F[i - j - 1] * tr[j]
        P *= (j - i)
    F[i] += P * tr[i]
    FT *= -i - 1
    R[N - i - 1] = F[i] / FT

for i in range(N + 1):
    R[i] = R[i].simplify()

display(Math(r'R = %s'%latex(R)))
```

```python
sum = Matrix.zeros(N, N)
C = Matrix.eye(N)
for i in range(N + 1):
    sum += R[i] * C
    C = mat * C

display(Math(r'\sum_{i=0}^{%d} R_i A^i = %s'%(N, latex(sum))))
```

```python
for i in range(N):
    for j in range(N):
        sum[i,j] = sum[i,j].simplify()

display(Math(r'\sum_{i=0}^{%d} R_i A^i = %s'%(N, latex(sum))))
```
